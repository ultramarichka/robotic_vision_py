```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import matplotlib.cm as cm
```

```python
def loadImageRaw(iPath, iSize, iFormat):
  #a=np.fromfile("/home/maryna/Documents/programming/ziga/RV/materials/1/slika-8bit.raw", dtype='uint8')
  a = np.fromfile(iPath, dtype=iFormat)
  b = np.reshape(a, iSize) 
  return b
```

```python
def showImage(iImage, iTitle):
  plt.imshow(iImage, cmap = cm.Greys_r)
  plt.gca().set_aspect('equal', 'datalim')
  plt.title(iTitle)
```

```python
showImage(oImage, 'Kukuc')
```
