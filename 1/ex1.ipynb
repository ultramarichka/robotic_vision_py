```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y        
        
    def l(self):
        return np.sqrt(self.x**2 + self.y**2)
    
    def dot(self, a):
        return self.x*a.x + self.y*a.y
```

```python
a = Vector(5,11)
```
